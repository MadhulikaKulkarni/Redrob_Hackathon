```python
import json

candidates = []

with open("../data/candidates.jsonl", "r", encoding="utf8") as f:
    for line in f:
        candidates.append(json.loads(line))

print("Total candidates:", len(candidates))
```

```python
from collections import Counter

titles = []

for c in candidates:
    titles.append(
        c["profile"]["current_title"]
    )

Counter(titles).most_common(30)
```

```python
all_skills = []

for c in candidates:
    for s in c["skills"]:
        all_skills.append(
            s["name"]
        )

Counter(all_skills).most_common(50)
```

```python
industries = []

for c in candidates:
    industries.append(
        c["profile"]["current_industry"]
    )

Counter(industries).most_common(20)
```
